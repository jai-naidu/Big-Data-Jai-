# Part 1 Analysis

    With this code we were able to analyze the patterns of appearance of 6 different birds, by month, in the King County. In order to achieve this task, I had to sift through a lot of unnecessary data from ebird. There was a lot of interesting data in the csv files including specific location of the observations of the different birds, longitude and latitude of the bird sightings, the list goes on. However, the data that was necessary for this part of the lab was the specific date of the observations. Specifically, I needed to know when the observations occurred so I could arrange the observations by month. I programmed the code to go through all of the observation dates and pinpoint what month the observations took place on (I had it check the first part of the date i.e. "11/..." would be November). If the code saw the date 11/12/15, it would see that the observation took place in November and it would add 1 to the "November" variable. This was done for every bird, for every month of the year. Then all the data was printed in a readable format.
    We learned a few things from the data. First of all, some birds did not appear at all in certain months. For example, the Snowy Owl only appeared in November, December, January, February, and March; the Bramblings only appeared in February, March, and April; and the Black Swift only appeared in May, June, April, August, and September. However, the Harlequin Ducks and Anna's Hummingbirds appeared every day of the month. These patterns have some implication. From these patterns we can assume a few things for each of these birds: seasonal preference, temperature preference, density of the population, and maybe even migration patterns (more data in different counties/ states would help). For instance, the Snowy Owls only show up during the Winter season, implying they prefer the cold weather; the Lazuli Bunting on the other hand appeared only in the Summer season. This pattern of appearance carries over to the Black Swift a little. The Black Swifts appear during late summer and stop appearing in the early winter. As for the Bramblings, they only appear 3 months of the year (February, March, and April). This could mean a few things: The Bramblings pass through the King County during this time of year, something disrupted their migration/general path and pushed them off-course into the King county, there are a lot of possibilities. Lastly, Anna's Hummingbirds seem to be very common relative to the other birds. There are thousands of sightings every month with no exception, implying that they are a plentiful species in the area. However, the numbers do seem disproportionate.
    There are a few things that must be taken into account when approaching this data. Since this is an open source data center, the data is not completely consistent. For instance, there is a possibility that more people observe Anna's Hummingbird over other species. In this case, it is quite evident that there is far more data on Anna's Hummingbird than the other birds. This causes the data to look like Anna's Hummingbird is FAR more prevalent than other species, while this might not be true. There is also a possibility that people's observations overlap. Two people might see the same bird.

```python
#FINDING THE AVERAGE NUMBER OF BIRDS PER OBSERVATION

observationlista=[]
for row in alist:
    observationlista.append(int(row[7]))   
print("Average of Snowy Owls per Observation", statistics.mean(observationlista))

observationlistb=[]
for row in blist:
    observationlistb.append(int(row[7]))   
print("Average of Lazuli Buntings per Observation", statistics.mean(observationlistb))

observationlistc=[]
for row in clist:
    observationlistc.append(int(row[7]))    
print("Average of Harlequin Ducks per Observation", statistics.mean(observationlistc))

observationlistd=[]
for row in dlist:
    observationlistd.append(int(row[7]))    
print("Average of Bramblings per Observation", statistics.mean(observationlistd))

observationliste=[]
for row in elist:
    observationliste.append(int(row[7]))  
print("Average of Black Swifts per Observation", statistics.mean(observationliste))

observationlistf=[]
for row in flist:
    observationlistf.append(int(row[7])) 
print("Average of Anna's Hummingbird per Observation", statistics.mean(observationlistf))

#GRAPH

AverageBirds = []

AverageBirds.append (statistics.mean(observationlista))
AverageBirds.append (statistics.mean(observationlistb))
AverageBirds.append (statistics.mean(observationlistc))
AverageBirds.append (statistics.mean(observationlistd))
AverageBirds.append (statistics.mean(observationliste))
AverageBirds.append (statistics.mean(observationlistf))

plt.hist(AverageBirds)
plt.title("Average Number of Bird Sightings Per Observation")
plt.ylabel("Frequency")
plt.xlabel("Average Number of Bird Sightings")
```
