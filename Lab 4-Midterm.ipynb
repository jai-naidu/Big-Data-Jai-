```python
import statistics
import numpy as np
import scipy.stats as st
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
filea = open("college majors.csv", "r", errors = "ignore")
alist = []
for i in filea:
    alist.append(i)
del alist[0]
for j in range (len(alist)):
    alist[j] = alist[j].split(",")
    
fileb = open("recent-grads.csv", "r", errors = "ignore")
blist = []
for i in fileb:
    blist.append(i)
del blist[0]
for j in range (len(blist)):
    blist[j] = blist[j].split(",")
```

How does the unemployment rate vary throughout different college majors in the US(graduates ages 25+)?

```python
total = []   
employed = []
unemployed = []

alistlookup = {}

for i in range(len(alist)):
    if alist[i][2] not in alistlookup:
        alistlookup[alist[i][2]] = [float(alist[i][3]), float(alist[i][4]), float(alist[i][6])]
    else:
        alistlookup[alist[i][2]][0]+=float(alist[i][3])
        alistlookup[alist[i][2]][1]+=float(alist[i][4])
        alistlookup[alist[i][2]][2]+=float(alist[i][6])

for key in alistlookup:
    total.append(alistlookup[key][0])

for key in alistlookup:
    employed.append(alistlookup[key][1])
    
for key in alistlookup:
    unemployed.append(alistlookup[key][2])

k = alistlookup.keys()

print(alistlookup)
print(total)
print(employed)
print(unemployed)
```

```python
x = range(16)

plt.title("Employed and Uneployed Graduates (ages 25 and over)")
plt.xlabel("Major")
plt.ylabel("Number of Graduates")
plt.bar(x, employed, color = 'g')
plt.bar(x, unemployed, color = 'r', bottom = employed)
plt.show()
```

```python
regression = st.linregress(unemployed, total)
slope = regression [0]
yint = regression[1]
print(regression[0], regression[1])
plt.plot([0, 500000], [yint, slope*500000+yint])

print(st.linregress(unemployed, total))
plt.scatter(unemployed, total)
plt.title("The Correlation between Total and Unemployed College Graduates (age 25+)")
plt.ylabel("Number of College Graduates in Millions")
plt.xlabel("Number of Unemployed College Graduates")

print("")
print("The r value is 0.96311965975425884.")
```

How likely is a college graduate to go into a job that requires a college degree; a job that doesn't require a college degree; and a low-income job?

```python
cjob = []   
ncjob = []
lijob = []
jtotal = []
litjobs = []

jobs = {}

for i in range(len(blist)):
    if blist[i][3] not in jobs:
        jobs[blist[i][3]] = [float(blist[i][18]), float(blist[i][19]), float(blist[i][20]), float(blist[i][9])]
    else:
        jobs[blist[i][3]][0]+=float(blist[i][18])
        jobs[blist[i][3]][1]+=float(blist[i][19])
        jobs[blist[i][3]][2]+=float(blist[i][20])
        jobs[blist[i][3]][3]+=float(blist[i][9])

for key in jobs:
    cjob.append(jobs[key][0])

for key in jobs:
    ncjob.append(jobs[key][1])
    
for key in jobs:
    lijob.append(jobs[key][2])
    
for key in jobs:
    jtotal.append(jobs[key][3])
    
for key in jobs:
    litjobs.append(jobs[key][2]/jobs[key][3])

print(jobs)
print(cjob)
print(ncjob)
print(lijob)
print(jtotal)
print(litjobs)

k = blistlookup.keys()
```

# Analyisis

After observing the type of job that any given college graduate will get a few things come to light. First of all, looking at the low income percentages for each major, it is fairly clear that the more technical the field, the lower the unemployment rate is. This could very well be due to the fact that for technical fields you transition from school to work is more seamless (getting a job out of college is easier). This would be due to the fact that tech related jobs are very in demand. Moreover, if the technical majors are less popular it becomes that much easier to find a job. Lastly, very technical jobs, in general, seldom offer low pay jobs. All of these factors add up to a high percent of success in these fields. On the other hand, liberal arts fields, arts fields, and humanities fields, for example, seem to have a very high rate of students that go into low income jobs. This would likely be due to the difficulty of gaining recognition as an artist and the difficulty of finding a high-pay job with a liberal arts degrees or humanities degrees. This would likely be attributed to the lack of expensive services in such fields and/or the societal demand for a major. Generally speaking, the fields that lead to societal growth (fields that noticeably advance something) tend to have the least graduates going into low income jobs. For STEM graduates, it is entirely plausible that they have a higher chance of going into a high income job because their job is contributing to the technological boom that we are currently going through. Of course there are still exceptions like petroleum engineering (the overall most successful major), which isn't very progressive. But outliers always exist.
For 'no college degree required' jobs, there is nothing much to see. This data simply tells us which major categories have more jobs that don't require majors.
However, looking jobs that do require college degrees tells us something. To no surprise STEM field graduates mostly go into 'college degree required' jobs. This could be because of how companies look for employees. It is very possible that most companies want their employees to meet a certain standard, one of which is a college degree, which shows they are capable. For companies a college degree shows them that the employee will likely not need to be pampered in any way. Moreover, the education fields had a lot of 'college degree required' jobs. This is probably due to the difficulty of getting a teaching job without a degree. If you don't have a degree as an incoming teacher, it is hard to impress an employer with your resume. In contrast, business fields don't always require a degree (like entrepreneurship), so it isn't surprising that such a small amount of graduates go into a jobs that require degrees. Aside from that, these rates are fairly similar across other majors.
